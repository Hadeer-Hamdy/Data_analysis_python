```python
import pandas as pd
```

```python
london_data=pd.read_csv('London Housing Data.csv')
```

```python
london_data.head()
```

```python
london_data['date']=pd.to_datetime(london_data['date'])
london_data.info()
```

```python
london_data['year']=london_data['date'].dt.year
london_data
```

```python
london_data['month']=london_data['date'].dt.month
london_data
```

```python
no_crimes=london_data [london_data['no_of_crimes']==0]
no_crimes
```

```python
no_crimes.value_counts().sum()
```

```python
england_max_data=london_data[london_data['area']=='england']
england_max_data.groupby('year')['average_price'].max().reset_index()
```

```python
england_min_data=london_data [london_data['area']=='england']
england_min_data.groupby('year')['average_price'].min().reset_index()
```
